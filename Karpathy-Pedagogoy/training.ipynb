##### → Splits Dataset into Train, Test, and Validation:

```python
from sklearn.utils import shuffle

X, Y = shuffle(X, Y, random_state=101)


from sklearn.model_selection import train_test_split


train_ratio = 0.70
val_ratio   = 0.15
test_ratio  = 0.15


# Step 1: Split into train+val and test
X_temp, X_test, Y_temp, Y_test = train_test_split(X, Y, 
    test_size=test_ratio,
    random_state=101,
    shuffle=True,
)

val_relative_ratio = val_ratio / (train_ratio + val_ratio)

X_train, X_val, Y_train, Y_val = train_test_split(
    X_temp, Y_temp, test_size=val_relative_ratio,
    random_state=101,
    shuffle=True,
)

# Verify the sizes
print(f"Total samples: {len(X)}")
print(f"Training set: {len(X_train)} samples ({len(X_train)/len(X)*100:.1f}%)")
print(f"Validation set: {len(X_val)} samples ({len(X_val)/len(X)*100:.1f}%)")
print(f"Test set: {len(X_test)} samples ({len(X_test)/len(X)*100:.1f}%)")
```

##### → Cost Function—categorical cross entrorpy with Negative Log-likelihood:

```python
def softmax(logits):
    # When we have multiples outputs...
    elements = [logit.exp() for logit in logits]

    Σ = sum(elements) #unicode u→3a3
    
    return [(element/Σ) for element in elements]
#end-def

def NLL(ys, ps):
    return -sum([y*(p+0.00000000001).log() for y, p in zip(ys, ps)])
#end-def

def categorical_cross_entrorpy(Y, Yp):
    losses = []
    for ys, ysp in zip(Y, Yp):
        ps = softmax(ysp)
        losses.append(NLL(ys, ps))
    #end-for
    return sum(losses)/len(Y)
#end-def
```

##### → Training Loop:

```python
model = MLP(n_inputs=4, ns_neurons=[8, 3])

n_parameters = len(model.parameters())
print(n_parameters)
```

```python
n_epochs = 500+1
η = 1e-2   # learning rate
μ  = 0.99  # momentum rate
# λ = 1e-2 # weight decay


losses_train = []
losses_val   = []

for epoch in range(n_epochs):
    # Forward pass:
    Y_train_prediction = [model(xs) for xs in X_train]
    loss_train = categorical_cross_entrorpy(Y_train, Y_train_prediction)

    Y_val_prediction = [model(xs) for xs in X_val]
    loss_val   = categorical_cross_entrorpy(Y_val, Y_val_prediction)

    # Backward pass:
    model.zero_grad()
    loss_train.backward()

    # Update parameters—vanilla gradients:
    # for parameter in model.parameters():
    #     parameter.data -= η * parameter.grad
    # #end-for

    # # Update parameters—momentum gradients:
    # for parameter in model.parameters():
    #     parameter.velocity = (μ * parameter.velocity) - (η * parameter.grad)
    #     parameter.data += parameter.velocity
    # #end-for

    # Update parameters—Nesterov momentum gradients:
    for parameter in model.parameters():
        v_cached = parameter.velocity
        parameter.velocity = (μ * parameter.velocity) - (η * parameter.grad)
        parameter.data += (-μ * v_cached) + ((1.0+μ) * parameter.velocity)
    #end-for


    if epoch % 25 == 0:
        print(f'When epoch is {epoch}, training loss is {loss_train.data:0.5f} and validation loss is {loss_val.data:0.5f}.')
    #end-if/else

    losses_train.append(loss_train.data)
    losses_val.append(loss_val.data)
#end-for
```

##### → Visualize the loss plot between `training loss` and `validation loss`:

```python
def loss_plot(losses_train, losses_val):
    plt.title('How does loss function change over training steps?', fontsize=12)

    plt.plot(losses_train, color='red', marker='.', linewidth=2.0, label='Training Loss')
    plt.plot(losses_val, color='green', marker='+', linewidth=2.0,  label='Validation Loss')

    plt.xlabel('Epoch (Training Step)', fontsize=10)
    plt.ylabel('Negative Log-Likelihood Loss', fontsize=10)

    plt.legend()

    plt.savefig(fname='Nesterov_Momentum_early_stopping_is_required.png', dpi=600)
    plt.show()
#end-def

loss_plot(losses_train, losses_val)
```

##### → Performance—Accuracy:

```python
def accuracy(X, Y):
    # n_classes = len(Y[0])
    Yp = [model(xs) for xs in X]

    ysp = []
    for yp in Yp:
        ps = softmax(yp) # [Value(data=0.4148), Value(data=0.2426), Value(data=0.3427]
        ysp.append(int(np.argmax([p.data for p in ps])))
    #end-for

    ys = []
    for y in Y:
        ys.append(int(np.argmax(y)))
    #end-for

    assert len(ys) == len(ysp)

    Σ = 0.0
    for y, yp in zip(ys, ysp):
        if y == yp:
            Σ += 1.0
        #end-if/else
    #end-for        

    return Σ/len(ys)
#end-def
    
value = accuracy(X_test, Y_test)
print(f'Test accuracy is {(value*100):0.2f}%.')
```
